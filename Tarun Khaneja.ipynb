# Box plot analysis

##### We will be identifying the outliers with the help of Boxplot

```python
import seaborn as sns

columns_to_check = ['loan_amnt', 'funded_amnt', 'funded_amnt_inv', 'installment', 'annual_inc', 'dti', 'revol_util']

# Boxplot to visualize outliers
for column in columns_to_check:
    sns.boxplot(y=df[column],color="cyan")
    plt.show()
```

```python
# As we see in above box plots for different metrices, there are many outliers present. Let's remove outliers.
```

```python
pd.options.display.float_format = '{:.2f}'.format 

def remove_outliers(metrices,df):
    for metric in metrices:
        print('Before Removal of Outliers :\n')
        print(df[metric].describe())

        # Data cleaning
        # Remove Outliers quantile .97 from metrices
        # it will make it easier to visualize the plots.

        df = df[df[metric] < df[metric].quantile(0.97)]

        print('After Removal of Outliers :')
        print(df[metric].describe())
        sns.boxplot(y=df[metric],color="green")
        plt.show()
    return df
    
df = remove_outliers(columns_to_check,df)
```

```python
# we have removed outliers, which can be seen in above box plots as well 
# and the effects of remove box plots is described with all statistics values
```

# Bivariate Analysis

```python
# Creating generic method for doing Bivariate analysis

def create_bivariate_plot( x_variable, y_variable, title, df, plot_type='bar'):
    plt.figure(figsize=(10, 6))
    
    if plot_type == 'bar':
        sns.barplot(x=x_variable, y=y_variable, data=df)
    elif plot_type == 'count':
        sns.countplot(x=x_variable, hue=y_variable, data=df)
    
    plt.title(title)
    plt.xlabel(x_variable)
    plt.ylabel(y_variable)
    plt.xticks(rotation=75)  # Rotate x-axis labels for better readability if needed
    plt.show()

 # Descriptive Statistics
    print("\nDescriptive Statistics for", x_variable, "and", y_variable, ":\n")
    print("Summary Statistics for", x_variable, ":\n", df[x_variable].describe())
    print("\nSummary Statistics for", y_variable, ":\n", df[y_variable].describe())
```

### comparing only with loan status (count) may not do correct analysis, because number of loans and charged off (Defaults) in particular categories are varying. Better way to calculate percentage for charged-off for each category

```python
# Note: Charged off are Defaults values
```

```python
#Bivariate Analysis on annual income against Chargedoff Percenatge :

annual_inc_vs_loan = df.groupby(['annual_inc_categories', 'loan_status']).loan_status.count().unstack().fillna(0).reset_index()
annual_inc_vs_loan['Total'] = annual_inc_vs_loan['Charged Off']+ annual_inc_vs_loan['Fully Paid']
annual_inc_vs_loan['Chargedoff_Percentage'] = (annual_inc_vs_loan['Charged Off'] / annual_inc_vs_loan['Total'])*100
annual_inc_vs_loan.sort_values('Chargedoff_Percentage', ascending=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is the name of your DataFrame
# Selecting only numerical columns for correlation matrix
numerical_cols = ['loan_amnt', 'funded_amnt', 'funded_amnt_inv', 'int_rate', 'installment', 'emp_length',
                  'annual_inc', 'dti', 'revol_util', 'pub_rec_bankruptcies']

# Creating a subset of numerical columns
numerical_data = df[numerical_cols]

# Calculating correlation matrix
correlation_matrix = numerical_data.corr()

# Plotting the correlation matrix using seaborn heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()

# Get statistical details
statistics = numerical_data.describe()
print(statistics)
```
